By harnessing this rich dataset, the Airbnb Data Analysis project aims to benefit the company and its stakeholders while enriching the experiences of millions of travelers and hosts worldwide.
````python
# REMOVE COLUMN THAT IS NOT USEFUL  OF MY OUTPUT AND ALSO CONTAINING MISSING VALUES

data = df.drop(['last_review'] , axis = 1)
````

````python
# TO SEE COLUMN IS REMOVE OR NOT

data.shape
````

````python
# TO CHECK HOW MANY MISSING VALUES PRESENT IN EACH COLUMN

print(data.isnull().sum())
````

````python
# TO FILL MISSING VALUES

data.fillna({'name' :  'not present' ,  'host_name' : 'absent' ,'reviews_per_month' : 0  }, inplace = True)
````

````python
# to check

data.isnull().sum()
````

````python
# Filtering discontinued listings and print

discontinued_listing = data[(data['price']==0 ) & (data['availability_365']==0)]
discontinued_listing
````

````python
# excludue data

data = data.loc[data['price']>0]
data
````
